```python
import json
import pandas as pd
import glob
```

```python
files = glob.glob("*.json")
# create a clank datafram 
df = pd.DataFrame()

for f in files:
    with open(f,'r') as f:
        for l in f:
            json_obj = json.loads(l)
            df = pd.concat([df,pd.DataFrame([json_obj])],ignore_index=True)

df
    
```

```python
import xml.etree.ElementTree as ET 

tree = ET.parse('source1.xml')
root = tree.getroot()
d = {"name":[], "height":[],"weight":[]}
for person in root:

    name = person.find("name").text
    weight = person.find("weight").text
    height = person.find("height").text
    d['name'].append(name)
    d['weight'].append(weight)
    d['height'].append(height)
```
