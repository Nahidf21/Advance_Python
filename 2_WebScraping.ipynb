```python
from bs4 import BeautifulSoup
import requests
import lxml
```

```python
url= "https://webscraper.io/test-sites/e-commerce/allinone/phones/touch"
page= requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text,"lxml")
soup
```

```python
#Find 
soup.find('header')
```

```python
soup.find('div',{'class':'navbar-header'})
```

```python
soup.find('h4', {'class':'price float-end card-title pull-right'})
```

```python
soup.find('h4', class_= "price float-end card-title pull-right")
```

```python
soup.find("h4", class_ = "price float-end card-title pull-right")
```

#### The find function only return the first value. If we want to return specific tag data then we need to use find_all function

```python
price = soup.find_all("h4", {'class': "price float-end card-title pull-right"} )
```

```python
print1 = soup.find_all("h4", {'class': "price float-end card-title pull-right"} )[1]
```

```python
name = soup.find_all('a', {'class': 'title'})
```

```python
review = soup.find_all('p', class_ = "review-count float-end")
```

```python
for i in range(0,len(price)):
    print(f"{name[i].get_text()}")
    print(f"Price:  {price[i].get_text()}")
    print(f"Review: {review[i].get_text()}")
    if int(review[i].get_text().split(' ')[0]) > 10 :
        print('Good Review')
    else:
        print("Avarage Review")
    print("\n")
```

## Find_all with multiple tags 


```python
soup.find_all(['h4','p'])
```

```python
soup.find_all(['h4','p','a'])
```

## Find_all with multiple tags and specific attributed

```python
soup.find_all(['h4','p','a'], id = True)
```

```python
soup.find_all('a', id = True)
```

```python
soup.find_all(id = True)
```

```python
soup.find_all(string="Iphone")
```

## Re libray can find similler string 

```python
import re
```

```python
soup.find_all(string= re.compile('Iph'))
```

```python
soup.find_all(string= re.compile('Nok'))
```

## We can use Re for find the attributes

```python
soup.find_all("h4", class_ = re.compile("pull"))
```

```python
soup.find_all(class_ = re.compile("pull"))
```

## limit the find_all function

```python
soup.find_all("h4",class_ = re.compile("pull"), limit= 4)
```

```python
soup.find_all('p', class_ = re.compile("description"))
```

## We can find list of string 

```python
soup.find_all(string= ["Iphone", "Nokia 123"])
```
